# Fix wrong processed_file_paths
in `/data/ongh0068/l1000/already_batched/valid_0/processed_file_paths.csv`, paths are like `/data/ongh0068/l1000/l1000_biaae/already_batched/valid_0/3_nsteps_1000.pkl.gz`, 
but there is no *l1000_biaae* folder any more

try to remove it

```python
import pandas as pd
```

```python
valid_f = '/data/ongh0068/l1000/already_batched/valid_0/wrong_processed_file_paths.csv'
valid_df = pd.read_csv(valid_f, header=0)
valid_df.head()
```

```python
c_filenames = [f.split('/l1000_biaae')[0]+f.split('/l1000_biaae')[1] for f in valid_df['file_names'].values]
c_filenames
```

```python
def correct_df(file):
    ori_df = pd.read_csv(file, header=0)
    c_filenames = [f.split('/l1000_biaae')[0]+f.split('/l1000_biaae')[1] for f in ori_df['file_names'].values]
    c_df = ori_df
    c_df['file_names'] = c_filenames
    
    return c_df
```

```python
c_valid_df = correct_df(valid_f)
c_valid_df.head()
```

```python
c_valid_df.to_csv('/data/ongh0068/l1000/already_batched/valid_0/processed_file_paths.csv', header=True, index=False)
```

## Guacamol data

```python
import os
```

```python
guacamol_path = '/data/conghao001/FYP/DrugDiscovery/data/guacamol/already_batched/'
dirs = os.listdir(guacamol_path)
dirs
```

```python
def correct_df_gua(file, split_name, data_path):
    ori_df = pd.read_csv(file, header=0)
    c_filenames = [os.path.join(data_path+split_name, f.split('/')[-1]) for f in ori_df['file_names'].values]
    c_df = ori_df
    c_df['file_names'] = c_filenames
    c_df.to_csv(file, header=True, index=False)
    
    return c_df
```

```python
for split_name in dirs:
    file = os.path.join(guacamol_path+split_name, 'processed_file_paths.csv')
    c_df = correct_df_gua(file, split_name, guacamol_path)
```
