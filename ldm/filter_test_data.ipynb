```python
import pandas as pd
```

```python
test_set_path = '/data/ongh0068/l1000/l1000_biaae/INPUT_DIR/test.csv'
ori_test_set = pd.read_csv(test_set_path)
ori_test_set
```

```python
test_set = ori_test_set.loc[((ori_test_set['CellLine']=='MCF7') | (ori_test_set['CellLine']=='VCAP')) 
        & (ori_test_set['Time']==24.0) 
        & ((ori_test_set['Dose']==5.0) | (ori_test_set['Dose']==10.0))]

test_set
```

```python
test_set = test_set.set_index('Unnamed: 0')
test_set.to_csv('filtered_test_set.csv', header=True)
# test_set.to_csv('filtered_test_set.csv', index=False)
```

```python
idxes = test_set['Unnamed: 0'].to_list()
len(idxes)
```

```python
l1000_df = pd.read_csv('/data/ongh0068/l1000/l1000_biaae/lincs/experiments_filtered.csv', index_col = False)
filtered_df = l1000_df.loc[l1000_df['Unnamed: 0'].isin(idxes)]
filtered_df
```
