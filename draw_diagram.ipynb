# Cond generation

```python
vae_path = 'ldm/cond_generation_res/ldm_vae_sim_scores.pkl'
aae_path = 'ldm/cond_generation_res/ldm_aae_sim_scores.pkl'
wae_path = 'ldm/cond_generation_res/ldm_wae_sim_scores.pkl'
```

```python
with open(vae_path, 'rb') as f:
    vae_ss = pickle.load(f)
    
with open(aae_path, 'rb') as f:
    aae_ss = pickle.load(f)
    
with open(wae_path, 'rb') as f:
    wae_ss = pickle.load(f)
```

```python
vae_ss
```

```python
maccs_ss = {}
for k, v in vae_ss.items():
    ref_smiles = k.split('_')[0]
    score = v['max_maccs_fp_tanimoto_sim'][0]
    smiles = v['max_maccs_fp_tanimoto_sim'][1]    
    
    if ref_smiles in maccs_ss.keys():
        maccs_ss[ref_smiles][smiles] = score
    else:
        maccs_ss[ref_smiles] = {}
        maccs_ss[ref_smiles][smiles] = score
        
```

```python
for k, v in aae_ss.items():
    ref_smiles = k.split('_')[0]
    score = v['max_maccs_fp_tanimoto_sim'][0]
    smiles = v['max_maccs_fp_tanimoto_sim'][1]    
    
    if ref_smiles in maccs_ss.keys():
        maccs_ss[ref_smiles][smiles] = score
    else:
        maccs_ss[ref_smiles] = {}
        maccs_ss[ref_smiles][smiles] = score
        
for k, v in wae_ss.items():
    ref_smiles = k.split('_')[0]
    score = v['max_maccs_fp_tanimoto_sim'][0]
    smiles = v['max_maccs_fp_tanimoto_sim'][1]    
    
    if ref_smiles in maccs_ss.keys():
        maccs_ss[ref_smiles][smiles] = score
    else:
        maccs_ss[ref_smiles] = {}
        maccs_ss[ref_smiles][smiles] = score
```

```python
def get_rand_n_sims(ss_dict, n=5):
    new_dict = {}
    for ref_smiles, sim_data in ss_dict.items():
        new_dict[ref_smiles] = {}
        keys = random.sample(sim_data.keys(), n)
#         values = [sim_data[k] for k in keys]
        for k in keys:
            new_dict[ref_smiles][k] = sim_data[k]
    return new_dict
```

```python
top5_ss = get_rand_n_sims(maccs_ss)
top5_ss
```

```python
img = {}
for k, v in top5_ss.items():
    smiles = [k] + list(v.keys())
    print(smiles)
    legends = ['Original inhibitor'] + [f'Similarity={ss:.2f}' for ss in v.values()]
    print(legends)
    mols = [Chem.MolFromSmiles(s) for s in smiles]
    img[k] = Draw.MolsToGridImage(mols, legends=legends, subImgSize=(250,250), molsPerRow=6)
    with open('molecule_imgs/cond_generation_100/'+k+'.png','wb+') as outf:
        outf.write(img[k].data)
```
